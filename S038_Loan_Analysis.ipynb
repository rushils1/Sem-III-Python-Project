```python
Highest_Loan=loan_dataset.nlargest(10,"LoanAmount")
Highest_Loan
```

Top 10 people having high loans


```python
Highest_Income=loan_dataset.nlargest(10,"ApplicantIncome")
Highest_Income
```

Top 10 people having high incomes

```python
intersection = pd.merge(Highest_Loan,Highest_Income,on=["Loan_ID","Gender","Married","Dependents","Education","Self_Employed","ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History","Property_Area","Loan_Status"],right_index=False)
intersection
```

There are 5 people who intersect in the high income and high loan criteria, this shows 50% of the sample population who have high loan and high income. This proves that higher the income higher the loan

```python
property_area=loan_dataset["Property_Area"].value_counts()
property_area
```

```python
sns.countplot(x='Property_Area', data=loan_dataset)
```

```python
Property_Area = pd.crosstab(loan_dataset['Property_Area'], loan_dataset['Loan_Status'])
Property_Area.div(Property_Area.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True, figsize=(6,5))
plt.legend(bbox_to_anchor=(1,1),loc='upper left', title='Loan_Status')
plt.show()
```

40% of people living in rural area have a loan on them

40% of people living in urban area have a loan on them

20% of people living in semiurban area have a loan on them

```python
Credit_History = pd.crosstab(loan_dataset['Credit_History'], loan_dataset['Loan_Status'])
Credit_History.div(Credit_History.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,figsize=(6,5))
plt.legend(bbox_to_anchor=(1,1),loc='upper left',title='Loan_Status')
plt.show()
```

Approximately 90% of people who dont have credit history dont have loan on them

Approximately 80% of people who have credit history have a loan on them
